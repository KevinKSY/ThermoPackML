```python
import pandas as pd
from numpy import *
%matplotlib inline

filename = '../tableRhsu.csv'
df = pd.read_csv('tableRhsu.csv')
```

```python
df
```

```python
# Define the neural network in tensorflows 
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor  
from sklearn.metrics import mean_squared_error

x_train, x_test, y_train, y_test = train_test_split(df.ix[:,['F','u','v ']].values, df['p'].values, random_state=0)
std_on_train = x_train.std(axis=0)
mean_on_train = x_train.mean(axis=0)
stdY_on_train = y_train.std(axis=0)
meanY_on_train = y_train.mean(axis=0)
print(std_on_train)
x_train_scaled = (x_train - mean_on_train) / std_on_train
x_test_scaled = (x_test - mean_on_train) / std_on_train
y_train_scaled = (y_train - meanY_on_train) / stdY_on_train
y_test_scaled = (y_test - meanY_on_train) / stdY_on_train

print("x_train shape: {}".format(x_train.shape))
print("x_test shape: {}".format(x_test.shape))
print("y_train shape: {}".format(y_train.shape))
print("y_train shape: {}".format(y_test.shape))
```

```python
thdyn_prop_df = pd.DataFrame(x_train,columns=['F','u','v'])
sm = pd.scatter_matrix(thdyn_prop_df, c=y_train, figsize=(10, 10), marker='o', hist_kwds={'bins': 20}, s=60, alpha=.8)
```

```python
dnn = MLPRegressor([100,100],activation='tanh')
dnn.fit(x_train_scaled, y_train_scaled)

print("Score (train): {:.2f}".format(dnn.score(x_train_scaled, y_train_scaled)))
print("Score (test): {:.2f}".format(dnn.score(x_test_scaled, y_test_scaled)))


from sklearn.externals import joblib
joblib.dump(dnn, 'dnnRegP_100_2_Fuv_tanh_scale')
```

```python
y_pred_train_scaled = dnn.predict(x_train_scaled)
y_pred_test_scaled = dnn.predict(x_test_scaled)
```

```python
y_pred_train = y_pred_train_scaled * stdY_on_train + meanY_on_train
y_pred_test = y_pred_test_scaled * stdY_on_train + meanY_on_train

errTrain = mean(abs(y_pred_train - y_train)/y_train)
errTest = mean(abs(y_pred_test - y_test)/y_test)

print("Score (train): {:.4f}".format(errTrain))
print("Score (test): {:.4f}".format(errTest))
```
